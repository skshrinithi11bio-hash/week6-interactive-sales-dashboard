```python
# ===============================
# INTERACTIVE SALES DASHBOARD
# ALL CHARTS IN ONE CELL
# ===============================

from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load dataset from Drive
csv_path = "/content/drive/MyDrive/arena files/sales_data.csv"
df = pd.read_csv(csv_path)

print(df.columns)

# Basic cleaning
df.fillna(0, inplace=True)

# Convert Date column if present
if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"])

plt.figure(figsize=(14,10))

# 1 Box Plot (Product vs Price)
plt.subplot(2,2,1)
sns.boxplot(x="Product", y="Price", data=df)
plt.title("Price Distribution by Product")

# 2 Violin Plot
plt.subplot(2,2,2)
sns.violinplot(x="Product", y="Price", data=df)
plt.title("Price Density by Product")

# 3 Line Plot (Date vs Price)
plt.subplot(2,2,3)
sns.lineplot(data=df, x="Date", y="Price")
plt.title("Sales Trend Over Time")

# 4 Heatmap
plt.subplot(2,2,4)
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")

plt.tight_layout()
plt.show()

# ===============================
# INTERACTIVE DASHBOARD (PLOTLY)
# ===============================

# 5 Interactive Bar Chart
fig1 = px.bar(df, x="Product", y="Price", color="Product",
              title="Interactive Sales by Product")
fig1.show()

# 6 Interactive Pie Chart
fig2 = px.pie(df, names="Product", values="Price",
              title="Revenue Share by Product")
fig2.show()
```
